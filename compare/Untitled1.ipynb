```python
import csv
import matplotlib.pyplot as plt
import networkx as nx
import warnings
warnings.filterwarnings('ignore') 
csv_reader = csv.reader(open("attachment3.csv"))
jiedian = []
all_con = []
for row in csv_reader:
    print(row)
    all_con.append(row)
    jiedian.append(row[0])
```

```python
all_tu = []
for i,j in zip(jiedian,all_con):
    all_tu.extend([(i,x) for x in j[1:]])
```

```python
all_tu
```

```python
import networkx as nx
g=nx.DiGraph()#创建空的有向图
g.add_edges_from(all_tu)
```

```python
g.edges
```

```python
nx.draw(g)

import matplotlib.pyplot as plt
plt.show()
```

```python
g = nx.cubical_graph()
nx.draw(g, pos=nx.spectral_layout(g), edge_color='b')
plt.show()
```
