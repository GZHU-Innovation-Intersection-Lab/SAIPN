```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

# Excel
df = pd.read_csv('matched_nodes_explicit_faster2.csv')
df['yinshi_w_value'] = pd.to_datetime(df['yinshi_w_value'], errors='coerce')
df['xianshi_w_value'] = pd.to_datetime(df['xianshi_w_value'], errors='coerce')

# 
df_yinshi = df[['yinshi_node_id', 'yinshi_w_value']].rename(columns={'yinshi_node_id': 'node_id', 'yinshi_w_value': 'date'})
df_xianshi = df[['xianshi_node_id', 'xianshi_w_value']].rename(columns={'xianshi_node_id': 'node_id', 'xianshi_w_value': 'date'})
df_yinshi['network_type'] = 'implict network'
df_xianshi['network_type'] = 'explict network'

# 
df_combined = pd.concat([df_yinshi, df_xianshi])

# 
df_combined['year_month'] = df_combined['date'].dt.to_period('M')

# Seaborn
sns.set(style="whitegrid")

# 
plt.figure(figsize=(15, 8))
ax = sns.pointplot(x=df_combined.index, y='year_month', hue='network_type', data=df_combined,
                   palette={"implict network": "blue", "explict network": "orange"}, markers=["o", "x"], linestyles=["-", "--"])

# 
ax.set_title('Optimized matched_nodes_explicit_faster2')
ax.set_ylabel('Year-Month')
ax.set_xlabel('Node Index')
plt.legend(title='Network Type', loc='upper right')

# 
plt.tight_layout()
plt.savefig(r'D:\academy\research\graphkun\compare\explicit_faster2.png', bbox_inches='tight')
plt.show()
```
