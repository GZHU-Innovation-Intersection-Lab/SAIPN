```python
import pandas as pd
import ast  # 更安全地解析字符串为列表

def load_data():
    # 读取CSV文件
    df_yinshi = pd.read_csv('top10_percent_yinshi.csv')
    df_xianshi = pd.read_csv('top10_percent_xianshi.csv')
    edge_log_df = pd.read_csv('edge_log.csv')
    return df_yinshi, df_xianshi, edge_log_df

def build_yinshi_dict(df_yinshi):
    """构建隐式网络字典，记录各节点最小传播时间"""
    yinshi_data = []
    for _, row in df_yinshi.iterrows():
        try:
            nodes = ast.literal_eval(row['top10_percent_nodes'])
        except:
            nodes = []
        for node_id, _ in nodes:
            yinshi_data.append({'node_id': node_id, 'w_value': row['w_value']})
    
    yinshi_df = pd.DataFrame(yinshi_data)
    return yinshi_df.groupby('node_id')['w_value'].min().to_dict()

def process_explicit_nodes(df_xianshi, yinshi_dict):
    """处理显式网络节点，生成包含隐式时间的DataFrame"""
    xianshi_data = []
    for _, row in df_xianshi.iterrows():
        try:
            nodes = ast.literal_eval(row['top10_percent_nodes'])
        except:
            nodes = []
        for node_id, _ in nodes:
            xianshi_data.append({
                'node_id': node_id,
                'w_value_xianshi': row['w_value']
            })
    
    xianshi_df = pd.DataFrame(xianshi_data)
    xianshi_df['w_value_yinshi'] = xianshi_df['node_id'].map(yinshi_dict)
    return xianshi_df

def process_related_nodes(edge_log_df):
    """预处理边日志，构建目标节点到源节点的映射"""
    return edge_log_df.groupby('Target_Node')['Source_Node'].agg(list).to_dict()

def main():
    df_yinshi, df_xianshi, edge_log_df = load_data()
    yinshi_dict = build_yinshi_dict(df_yinshi)
    xianshi_df = process_explicit_nodes(df_xianshi, yinshi_dict)
    edge_dict = process_related_nodes(edge_log_df)
    
    # 分割存在/不存在隐式节点的显式节点
    mask = xianshi_df['w_value_yinshi'].notna()
    df_exists, df_not_exists = xianshi_df[mask], xianshi_df[~mask]
    
    # 处理存在的节点
    implicit_exists = df_exists[df_exists['w_value_yinshi'] < df_exists['w_value_xianshi']]
    explicit_exists = df_exists[df_exists['w_value_yinshi'] > df_exists['w_value_xianshi']]
    
    # 处理不存在的节点
    implicit_not_exists, explicit_not_exists = [], []
    for _, row in df_not_exists.iterrows():
        node_id, wx = row['node_id'], row['w_value_xianshi']
        sources = edge_dict.get(node_id, [])
        
        # 获取所有关联隐式节点的时间
        related_times = [yinshi_dict.get(src, float('inf')) for src in sources]
        valid_times = [t for t in related_times if t != float('inf')]
        
        if not valid_times:
            continue
        
        min_time, max_time = min(valid_times), max(valid_times)
        if min_time < wx:
            implicit_not_exists.extend([
                {'node_id': src, 'yinshi_w_value': t, 'xianshi_w_value': wx, 'related_xianshi_node': node_id}
                for src, t in zip(sources, related_times) if t < wx
            ])
        elif max_time > wx:
            explicit_not_exists.extend([
                {'node_id': src, 'yinshi_w_value': t, 'xianshi_w_value': wx, 'related_xianshi_node': node_id}
                for src, t in zip(sources, related_times) if t > wx
            ])
    
    # 合并结果
    implicit_nodes = pd.DataFrame(implicit_exists[['node_id', 'w_value_xianshi', 'w_value_yinshi']]
                                  ).rename(columns={'w_value_yinshi': 'yinshi_w_value', 
                                                  'w_value_xianshi': 'xianshi_w_value'})
    implicit_nodes = pd.concat([implicit_nodes, pd.DataFrame(implicit_not_exists)])
    
    explicit_nodes = pd.DataFrame(explicit_exists[['node_id', 'w_value_xianshi', 'w_value_yinshi']]
                                  ).rename(columns={'w_value_yinshi': 'yinshi_w_value',
                                                  'w_value_xianshi': 'xianshi_w_value'})
    explicit_nodes = pd.concat([explicit_nodes, pd.DataFrame(explicit_not_exists)])
    
    # 统计结果
    total = len(xianshi_df)
    implicit_count = len(implicit_nodes)
    explicit_count = len(explicit_nodes)
    
    print(f"隐式网络更快的节点数: {implicit_count}, 占比: {implicit_count/total:.2%}")
    print(f"显式网络更快的节点数: {explicit_count}, 占比: {explicit_count/total:.2%}")
    
    # 保存结果
    implicit_nodes.to_csv('implicit_faster.csv', index=False)
    explicit_nodes.to_csv('explicit_faster.csv', index=False)

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import ast  # 更安全地解析字符串为列表

def load_data():
    """加载数据文件"""
    df_yinshi = pd.read_csv('top10_percent_yinshi.csv')
    df_xianshi = pd.read_csv('top10_percent_xianshi.csv')
    edge_log_df = pd.read_csv('edge_log.csv')
    return df_yinshi, df_xianshi, edge_log_df

def build_yinshi_dict(df_yinshi):
    """构建隐式网络字典，记录各节点的最小传播时间"""
    yinshi_dict = {}
    for _, row in df_yinshi.iterrows():
        try:
            nodes = ast.literal_eval(row['top10_percent_nodes'])
        except (ValueError, SyntaxError):
            nodes = []
        for node_id, _ in nodes:
            if node_id not in yinshi_dict or row['w_value'] < yinshi_dict[node_id]:
                yinshi_dict[node_id] = row['w_value']
    return yinshi_dict

def build_edge_dict(edge_log_df):
    """构建边日志字典，记录每个目标节点对应的源节点列表"""
    edge_dict = {}
    for _, row in edge_log_df.iterrows():
        target = row['Target_Node']
        source = row['Source_Node']
        if target not in edge_dict:
            edge_dict[target] = []
        edge_dict[target].append(source)
    return edge_dict

def compare_propagation_times(df_xianshi, yinshi_dict, edge_dict):
    """比较隐式和显式网络中节点的传播时间"""
    implicit_faster_nodes = []
    explicit_faster_nodes = []
    implicit_faster_count = 0
    explicit_faster_count = 0
    total_check_count = 0

    # 用于记录已经处理过的隐式节点，避免重复统计
    processed_implicit_nodes = set()

    for _, row in df_xianshi.iterrows():
        try:
            nodes = ast.literal_eval(row['top10_percent_nodes'])
        except (ValueError, SyntaxError):
            nodes = []
        w_value_xianshi = row['w_value']

        for xianshi_node_id, _ in nodes:
            total_check_count += 1

            # 直接比较隐式和显式时间
            if xianshi_node_id in yinshi_dict:
                w_value_yinshi = yinshi_dict[xianshi_node_id]
                if w_value_yinshi < w_value_xianshi:
                    implicit_faster_count += 1
                    implicit_faster_nodes.append({
                        'node_id': xianshi_node_id,
                        'yinshi_w_value': w_value_yinshi,
                        'xianshi_w_value': w_value_xianshi
                    })
                elif w_value_yinshi > w_value_xianshi:
                    explicit_faster_count += 1
                    explicit_faster_nodes.append({
                        'node_id': xianshi_node_id,
                        'xianshi_w_value': w_value_xianshi,
                        'yinshi_w_value': w_value_yinshi
                    })
            else:
                # 检查关联节点
                if xianshi_node_id in edge_dict:
                    for related_node_id in edge_dict[xianshi_node_id]:
                        if related_node_id in yinshi_dict and related_node_id not in processed_implicit_nodes:
                            w_value_yinshi = yinshi_dict[related_node_id]
                            if w_value_yinshi < w_value_xianshi:
                                implicit_faster_count += 1
                                implicit_faster_nodes.append({
                                    'node_id': related_node_id,
                                    'yinshi_w_value': w_value_yinshi,
                                    'xianshi_w_value': w_value_xianshi,
                                    'related_xianshi_node': xianshi_node_id
                                })
                                processed_implicit_nodes.add(related_node_id)
                            elif w_value_yinshi > w_value_xianshi:
                                explicit_faster_count += 1
                                explicit_faster_nodes.append({
                                    'node_id': related_node_id,
                                    'xianshi_w_value': w_value_xianshi,
                                    'yinshi_w_value': w_value_yinshi,
                                    'related_xianshi_node': xianshi_node_id
                                })
                                processed_implicit_nodes.add(related_node_id)

    return implicit_faster_nodes, explicit_faster_nodes, implicit_faster_count, explicit_faster_count, total_check_count

def main():
    # 加载数据
    df_yinshi, df_xianshi, edge_log_df = load_data()

    # 构建隐式网络字典和边日志字典
    yinshi_dict = build_yinshi_dict(df_yinshi)
    edge_dict = build_edge_dict(edge_log_df)

    # 比较传播时间
    implicit_faster_nodes, explicit_faster_nodes, implicit_faster_count, explicit_faster_count, total_check_count = compare_propagation_times(df_xianshi, yinshi_dict, edge_dict)

    # 计算比例
    implicit_faster_ratio = implicit_faster_count / total_check_count if total_check_count > 0 else 0
    explicit_faster_ratio = explicit_faster_count / total_check_count if total_check_count > 0 else 0

    # 保存结果
    pd.DataFrame(implicit_faster_nodes).to_csv('matched_nodes_implicit_faster.csv', index=False, encoding='utf-8')
    pd.DataFrame(explicit_faster_nodes).to_csv('matched_nodes_explicit_faster.csv', index=False, encoding='utf-8')

    # 打印结果
    print(f"隐式网络比显式网络快的频次: {implicit_faster_count}, 比例: {implicit_faster_ratio:.2%}")
    print(f"显式网络比隐式网络快的频次: {explicit_faster_count}, 比例: {explicit_faster_ratio:.2%}")

if __name__ == "__main__":
    main()
```

```python
import pandas as pd

# 读取 top5_yinshi.csv 和 top5_xianshi.csv
df_xianshi =pd.read_csv('top10_percent_xianshi.csv')
df_yinshi = pd.read_csv('top10_percent_yinshi.csv')

xianshi_dict = {}

# 遍历 top5_xianshi.csv，将节点 ID 和对应的 w_value 存储到字典中
for index, row in df_xianshi.iterrows():
    w_value = row['w_value']
    top5_nodes = eval(row['top10_percent_nodes'])  # 解析节点列表
    for node_id, _ in top5_nodes:
        if node_id not in xianshi_dict:
            xianshi_dict[node_id] = w_value

# 读取 edge_log.csv 文件，存储每一条边的连接信息
edge_log_df = pd.read_csv('edge_log.csv')

# 用于存储符合条件的节点
matched_nodes = []

# 计数器：统计显式网络比隐式网络快的频次
explicit_faster_count = 0
total_check_count = 0  # 用于计算比例

# 遍历 top5_yinshi.csv，检查每个节点是否存在于 xianshi_dict 且时间更晚，或者在 edge_log.csv 中有连接关系
for index, row in df_yinshi.iterrows():
    w_value_yinshi = row['w_value']
    top5_nodes = eval(row['top10_percent_nodes'])  # 解析节点列表
    for node_id, _ in top5_nodes:
        total_check_count += 1  # 每检查一个节点，计数增加

        # 用来记录是否找到显式网络节点的时间既比隐式网络早又比隐式网络晚
        found_earlier = False
        found_later = False

        # 如果该节点在隐式网络中
        if node_id in xianshi_dict:
            w_value_xianshi = xianshi_dict[node_id]
            if w_value_yinshi < w_value_xianshi:  # 判断xian式网络节点是否更晚
                found_later = True
            elif w_value_yinshi > w_value_xianshi:  # 判断显式网络节点是否更早
                found_earlier = True

        # 如果没有同时找到比显式网络节点早和晚的情况，则继续检查
        if not (found_earlier and found_later):
            # 如果找到显式网络节点比隐式网络节点早的情况
            if found_earlier:
                matched_nodes.append({
                    'xianshi_node_id': node_id,
                    'xianshi_w_value': w_value_xianshi,
                    'yinshi_node_id': node_id,
                    'yinshi_w_value': w_value_yinshi
                })
                explicit_faster_count += 1
            # 如果该节点在 edge_log.csv 中有连接关系，继续查找
            elif not found_earlier and not found_later:
                related_nodes = edge_log_df[
                    (edge_log_df['Target_Node'] == node_id)]  # 确保显式网络节点是起始点
                for _, edge_row in related_nodes.iterrows():
                    # 取得与该节点相关的另一节点的ID（隐式网络节点）
                    related_node_id = edge_row['Source_Node']

                    if related_node_id in xianshi_dict:
                        w_value_xianshi = xianshi_dict[related_node_id]
                        if w_value_yinshi > w_value_xianshi:  # 显式网络的时间更早
                            found_earlier = True
                            matched_nodes.append({
                                'xianshi_node_id': related_node_id,
                                'xianshi_w_value': w_value_xianshi,
                                'yinshi_node_id':  node_id,
                                'yinshi_w_value': w_value_yinshi
                            })
                            explicit_faster_count += 1
                            break  # 只处理第一次匹配，跳出相关节点循环
                        elif w_value_yinshi < w_value_xianshi:  # 隐式网络的时间更晚
                            found_later = True
                            break  # 找到一次晚于显式网络节点，跳出
                # 如果同时发现了比隐式网络节点时间早和比隐式网络节点时间晚的情况，跳过
                if found_earlier and found_later:
                    continue  # 跳过该节点，继续遍历下一个

# 将符合条件的节点存储到一个 DataFrame 中
matched_df = pd.DataFrame(matched_nodes)

# 计算比例
explicit_faster_ratio = explicit_faster_count / total_check_count if total_check_count > 0 else 0

# 保存符合条件的节点到 CSV 文件
matched_df.to_csv('matched_nodes_explicit_faster2.csv', index=False, encoding='utf-8')

# 打印结果
print(f"符合条件的节点已保存到 'matched_nodes_explicit_faster2.csv'！")
print(f"显式网络比隐式网络快的频次: {explicit_faster_count}")
print(f"显式网络比隐式网络快的比例: {explicit_faster_ratio:.2%}")
```

```python
import pandas as pd

# 读取隐式和显式网络节点数据
df_yinshi = pd.read_csv('top10_percent_yinshi.csv')
df_xianshi = pd.read_csv('top10_percent_xianshi.csv')

# 将隐式网络节点及其传播时间存储到字典中
yinshi_dict = {}
for _, row in df_yinshi.iterrows():
    w_value_yinshi = row['w_value']
    yinshi_nodes = eval(row['top10_percent_nodes'])  # 假设节点列表是字符串格式
    for yinshi_node_id, _ in yinshi_nodes:
        if yinshi_node_id not in yinshi_dict:
            yinshi_dict[yinshi_node_id] = w_value_yinshi

# 将显式网络节点及其传播时间存储到字典中
xianshi_dict = {}
for _, row in df_xianshi.iterrows():
    w_value_xianshi = row['w_value']
    xianshi_nodes = eval(row['top10_percent_nodes'])  # 假设节点列表是字符串格式
    for xianshi_node_id, _ in xianshi_nodes:
        if xianshi_node_id not in xianshi_dict:
            xianshi_dict[xianshi_node_id] = w_value_xianshi

# 读取边日志文件，构建邻接表
edge_log_df = pd.read_csv('edge_log.csv')
adjacency_list = {}
for _, row in edge_log_df.iterrows():
    source_node = row['Source_Node']
    target_node = row['Target_Node']
    if source_node not in adjacency_list:
        adjacency_list[source_node] = []
    if target_node not in adjacency_list:
        adjacency_list[target_node] = []
    adjacency_list[source_node].append(target_node)
    adjacency_list[target_node].append(source_node)

# 用于存储符合条件的节点
matched_nodes = []

# 统计变量
explicit_faster_count = 0
implicit_faster_count = 0
total_check_count = 0

# 遍历显式网络节点，检查传播时间
for xianshi_node_id, w_value_xianshi in xianshi_dict.items():
    total_check_count += 1
    found_earlier = False
    found_later = False

    # 检查显式网络节点是否直接存在于隐式网络中
    if xianshi_node_id in yinshi_dict:
        w_value_yinshi = yinshi_dict[xianshi_node_id]
        if w_value_xianshi < w_value_yinshi:
            found_earlier = True
        elif w_value_xianshi > w_value_yinshi:
            found_later = True

    # 如果未找到直接匹配，检查关联节点
    if not found_earlier and not found_later:
        if xianshi_node_id in adjacency_list:
            for related_node_id in adjacency_list[xianshi_node_id]:
                if related_node_id in yinshi_dict:
                    w_value_yinshi = yinshi_dict[related_node_id]
                    if w_value_xianshi < w_value_yinshi:
                        found_earlier = True
                        break
                    elif w_value_xianshi > w_value_yinshi:
                        found_later = True
                        break

    # 根据统计结果记录信息
    if found_earlier:
        explicit_faster_count += 1
        matched_nodes.append({
            'xianshi_node_id': xianshi_node_id,
            'xianshi_w_value': w_value_xianshi,
            'yinshi_node_id': xianshi_node_id if xianshi_node_id in yinshi_dict else related_node_id,
            'yinshi_w_value': yinshi_dict[xianshi_node_id] if xianshi_node_id in yinshi_dict else w_value_yinshi
        })
    elif found_later:
        implicit_faster_count += 1

# 将符合条件的节点保存为 DataFrame
matched_df = pd.DataFrame(matched_nodes)

# 计算比例
explicit_faster_ratio = explicit_faster_count / total_check_count if total_check_count > 0 else 0
implicit_faster_ratio = implicit_faster_count / total_check_count if total_check_count > 0 else 0

# 保存结果到 CSV 文件
matched_df.to_csv('matched_nodes_explicit_faster_kimi1.csv', index=False, encoding='utf-8')

# 打印结果
print(f"符合条件的节点已保存到 'matched_nodes_explicit_faster.csv'！")
print(f"显式网络比隐式网络快的频次: {explicit_faster_count}")
print(f"显式网络比隐式网络快的比例: {explicit_faster_ratio:.2%}")
print(f"隐式网络比显式网络快的频次: {implicit_faster_count}")
print(f"隐式网络比显式网络快的比例: {implicit_faster_ratio:.2%}")
```
