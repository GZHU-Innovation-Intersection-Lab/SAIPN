```python
df1=pd.read_excel('附件3-change.xlsx',sheet_name='Sheet3')
df2=pd.read_excel('附件3-change.xlsx',sheet_name='交通路口路线数据')
li=list(df1['节点编号'])
df=df2[(df2['路线起点'].isin(li))& (df2['路线终点'].isin(li))]
qi=list(df['路线起点'])
zhong=list(df['路线终点'])
quan=list(df['路线距离(m)'])
nodes=[]
for i in range(len(qi)):
    nodes.append((qi[i],zhong[i],quan[i]))
key=[]
nodes_location=[]
for i in range(len(df1['节点编号'])):
    key.append(list(df1['节点编号'])[i])
    nodes_location.append((list(df1['路口横坐标'])[i],list(df1['路口纵坐标'])[i]))
dic=dict(zip(key,nodes_location))
```

```python
import networkx as nx  # 导入 NetworkX 工具包
gAnt = nx.Graph()  # 创建：空的 无向图
gAnt.add_weighted_edges_from(nodes)
pos=dic
fig, ax = plt.subplots(figsize=(18, 12))
nx.draw(gAnt, pos, with_labels=False, node_color='#000080', alpha=0.6,node_size=100,node_shape='^',width=0.7)
labels = nx.get_edge_attributes(gAnt,'weight')  # 边的权值
nx.draw_networkx_edge_labels(gAnt,pos,edge_labels=labels, font_color='#000000',font_size=8,alpha=0.7)  # 显示边的权值
plt.savefig('networkx.png',dpi=450)
plt.show()
```
